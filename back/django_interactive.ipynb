```python
import csv
from app.controllers.payer_controller import PayerController
from pprint import pprint

file = open("media/547415fc-bcc7-4c8a-ada3-f05bd4c4f102/not_in_db.csv", "r", encoding="utf-8")
csv_reader = csv.reader(file, delimiter=',')

csv_reader.__next__()  # Pula o cabe√ßalho

not_found = []
found = []
for row in csv_reader:
    document = row[1]
    payer = PayerController.get(silent=True, user__cpf_cnpj__contains=document)
    if not payer:
        not_found.append(document)
    else:
        found.append(document)

print("Not found:")
pprint(not_found)
print("Found:")
pprint(found)
```
